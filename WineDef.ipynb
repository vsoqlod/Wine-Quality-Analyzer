### Correlation: positive vs negative

```python
correlation = wine_dataset.corr()
```

```python
# constructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='rainbow')
```

#### Color closer to red gives positive correlation, on the other hand, give negative correlation when closer to bluish color.

1. Alcohol and Quality: There's a positive correlation between alcohol content and wine quality, suggesting that higher alcohol may be associated with higher quality wines.

2. Acidity Attributes (Fixed Acidity, Volatile Acidity, Citric Acid): Fixed acidity has a positive correlation with citric acid but a negative correlation with volatile acidity, indicating that different types of acidity have varying impacts on wine characteristics.

3. pH and Acidity (Fixed Acidity, Citric Acid): pH shows a negative correlation with both fixed acidity and citric acid, consistent with the expected inverse relationship between acidity and pH level in wine.

4. Free Sulfur Dioxide and Total Sulfur Dioxide: There's a strong positive correlation between free sulfur dioxide and total sulfur dioxide, implying that free SO2 is a significant component of total SO2 in wine.

5. Density and Fixed Acidity: Density has a positive correlation with fixed acidity, suggesting that wines with higher acidity may have higher density.

### Data Preprocessing

```python
# seperate the data and label
X = wine_dataset.drop('quality', axis=1)
```

```python
print(X)
```

### Label Binarization : Standarize the value of the quality

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>= 7 else 0)
```

```python
print(Y)
```

### Train & Test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

### Model training: Random Forest

```python
model = RandomForestClassifier()
```

```python
# X = training data, Y = label values for training data (0,1)
model.fit(X_train, Y_train)
```

### Model Evaluation: Accuracy

#### Accuracy Score

```python
# accuracy on test data
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict, Y_test)
```

```python
print('Accuracy: ', test_data_accuracy)
```
